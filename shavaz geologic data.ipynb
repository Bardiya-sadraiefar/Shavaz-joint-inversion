# visualizing geologic data

```python
import pandas
df = pandas.read_csv("lil-13.csv")
print(df)
```

```python
import seaborn as sns
g = sns.PairGrid(df, hue="LITHOLOGY")
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot,hue=None,fill=True)
g.map_offdiag(sns.scatterplot, size=df["FeO"])
g.add_legend(title="", adjust_subtitles=True)
```
